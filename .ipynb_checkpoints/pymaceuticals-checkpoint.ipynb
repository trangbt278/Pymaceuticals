## Observations and Insights 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
data_complete = pd.merge(study_results, mouse_metadata, on = 'Mouse ID', how = 'left' )
# Display the data table for preview
data_complete
```

```python
# Checking the number of mice.
number_mice = len(data_complete['Mouse ID'].unique())
print (f'Number of mice in the original data: {number_mice}')
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicate_miceID = data_complete[data_complete.duplicated(subset = ['Mouse ID','Timepoint'], keep=False)]['Mouse ID'].unique()
for x in duplicate_miceID:
    print (f'Duplicate mice by ID: {x}')
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicate = data_complete[data_complete.duplicated(subset = ['Mouse ID','Timepoint'], keep=False)]
duplicate
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
for x in duplicate_miceID:
    data_clean = data_complete[data_complete['Mouse ID'] != x]
data_clean
```

```python
# Checking the number of mice in the clean DataFrame.
number_mice = len(data_clean['Mouse ID'].unique())
number_mice
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
groupby_regimen = data_clean.groupby('Drug Regimen')
mean_tumor = round(groupby_regimen['Tumor Volume (mm3)'].mean(), 2)
median_tumor = round(groupby_regimen['Tumor Volume (mm3)'].median(), 2)
var_tumor = round(groupby_regimen['Tumor Volume (mm3)'].var(), 2)
std_tumor = round(groupby_regimen['Tumor Volume (mm3)'].std(), 2)
sem_tumor = round(groupby_regimen['Tumor Volume (mm3)'].sem(), 2)
# Assemble the resulting series into a single summary dataframe.
summary_regimen = pd.DataFrame({'Mean of Tumor Volume (mm3) ': mean_tumor,
                               'Median of Tumor Volume (mm3)' : median_tumor,
                               'Varience of Tumor Volume (mm3)': var_tumor,
                               'STD of Tumor Volume (mm3)': std_tumor,
                               'SEM of Tumor Volume (mm3)': sem_tumor})
summary_regimen
```

## Bar and Pie Charts

```python
# Generate a bar plot showing the total number of timepoints for all mice tested for each drug regimen using Pandas.

#calculate total of timepoints for each regiment
total_timepoints = groupby_regimen['Timepoint'].sum()
#create a data frame of total number of timepoints vs drgu regiment
data_total_timepoints = pd.DataFrame(data = total_timepoints)
#reset index
data_total_timepoints = data_total_timepoints.reset_index()

#bar chart
ax = total_timepoints.plot.bar(x='Drug Regimen', y='Timepoint', color ='green')
#bar chart format
ax.set_xlabel('Drug Regiments')
ax.set_ylabel('Timepoints')
ax.set_title('Total number of timepoints vs Drug regiment - using Pandas')

```

```python
# Generate a bar plot showing the total number of timepoints for all mice tested for each drug regimen using pyplot.

#calculate x_pos for the bar chart based on list of Drug Regiment
x_pos = [i for i, _ in enumerate(data_total_timepoints['Drug Regimen'])]
#bar chart
plt.bar(x_pos, data_total_timepoints['Timepoint'], color = 'green')

#bar chart format
plt.xlabel('Drug Regiments')
plt.ylabel('Timepoints')
plt.title('Total number of timepoints vs Drug regiment - using Pylot')
plt.xticks(x_pos, data_total_timepoints['Drug Regimen'] )

plt.setp(plt.gca().get_xticklabels(), rotation=45, horizontalalignment='right')
plt.show()
```

```python
# Generate a pie plot showing the distribution of female versus male mice using Pandas
count_female = data_clean[data_clean['Sex'] == 'Female'].count()['Sex']
count_male = data_clean[data_clean['Sex'] == 'Male'].count()['Sex']

count_sex = pd.DataFrame({'sex_count': [count_female, count_male]}, index = ['Female', 'Male'])
plot = count_sex.plot.pie(y='sex_count', title='Female vs Male using Pandas', \
                          autopct='%1.1f%%', shadow=True)
count_sex
```

```python
# Generate a pie plot showing the distribution of female versus male mice using pyplot
count_sex = pd.DataFrame({'sex_count': [count_female, count_male], 'sex': ['Female', 'Male']})
fig1, ax1 = plt.subplots()
ax1.pie(count_sex['sex_count'], labels=count_sex['sex'], \
        autopct='%1.1f%%', shadow=True)
plt.title('Female vs Male using Pyplot')
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']
# Start by getting the last (greatest) timepoint for each mouse
greatest_tumor = data_clean.groupby('Mouse ID')['Timepoint'].max().reset_index()
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
final_tumor = greatest_tumor.merge(data_clean, on =['Mouse ID','Timepoint'], how = 'left')
final_tumor
```

```python
# Put treatments into a list for for loop (and later for plot labels)
# Create empty list to fill with tumor vol data (for plotting)
tumor_volumn_list = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for each in treatments:
    # Locate the rows which contain mice on each drug and get the tumor volumes
    final_tumor_volumn = final_tumor.loc[final_tumor['Drug Regimen'] == each, 'Tumor Volume (mm3)' ] 
    # add subset 
    tumor_volumn_list.append(final_tumor_volumn)
    # Determine outliers using upper and lower bounds
    quartiles = final_tumor_volumn.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    outlier = final_tumor_volumn.loc[(final_tumor_volumn < lower_bound) | (final_tumor_volumn > upper_bound)]
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
fig1, ax1 = plt.subplots()
ax1.set_title('Final Tumor (Capomulin, Ramicane, Infubinol, and Ceftamin)')
ax1.set_ylabel('Tumor Volume (mm3)')
ax1.set_xlabel('Drug Regiment')
ax1.boxplot(tumor_volumn_list)
plt.show()
```
